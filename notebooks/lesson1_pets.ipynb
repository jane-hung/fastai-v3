## Looking at the data

We are going to use the [Oxford-IIIT Pet Dataset](http://www.robots.ox.ac.uk/~vgg/data/pets/) by [O. M. Parkhi et al., 2012](http://www.robots.ox.ac.uk/~vgg/publications/2012/parkhi12a/parkhi12a.pdf) which features 12 cat breeds and 25 dogs breeds. Our model will need to learn to differentiate between these 37 distinct categories. According to their paper, the best accuracy they could get in 2012 was 59.21%, using a complex model that was specific to pet detection, with separate "Image", "Head", and "Body" models for the pet photos. Let's see how accurate we can be using deep learning!

We are going to use the `untar_data` function to which we must pass a URL as an argument and which will download and extract the data.

```
# finegrain classification to classify more granular labels/categories
help(untar_data)
# Union means fname can take a Path object or string object
# return pathlib.Path object
```

```
path = untar_data(URLs.PETS); path
```

```
# this path object can contain other paths
# didn't realize .ls() could work like this
path.ls()
```

```
# path objects! better than strings. use /
path_anno = path/'annotations'
path_img = path/'images'
```

The first thing we do when we approach a problem is to take a look at the data. We _always_ need to understand very well what the problem is and what the data looks like before we can figure out how to solve it. Taking a look at the data means understanding how the data directories are structured, what the labels are and what some sample images look like.

The main difference between the handling of image classification datasets is the way labels are stored. In this particular dataset, labels are stored in the filenames themselves. We will need to extract them to be able to classify the images into the correct categories. Fortunately, the fastai library has a handy function made exactly for this, `ImageDataBunch.from_name_re` gets the labels from the filenames using a [regular expression](https://docs.python.org/3.6/library/re.html).

```
# labels are stored in filenames and need to be extracted
fnames = get_image_files(path_img)
fnames[:5]
```

```
# set random seed
np.random.seed(2)
# set regular expressions to extract last words from the filenames. Start with / and end with a number + .jpg
pat = r'/([^/]+)_\d+.jpg$'
```

```
# ImageDataBunch contains all the data you need for the model
# size images want to work with. shortcut current deep learning tech. GPU has to apply same techniques to everything at the same time so can't do that if everything diff size
# size = 224 is normal
data = ImageDataBunch.from_name_re(path_img, fnames, pat, ds_tfms=get_transforms(), size=224, bs=bs
                                  ).normalize(imagenet_stats)

# ImageDataBunch contains train, valid, test and labels+images
# normalize to get mean = 0 and standard deviation = 1for each pixels (0-255)
# get_transform helps to center crop and grab middle bit and resize for size = 224; padding
```

```
data
```

```
data.show_batch(rows=3, figsize=(7,6))
```

```
print(data.classes)
# data bunch has attribute c which is the number of classes
len(data.classes),data.c
```

## Training: resnet34

Now we will start training our model. We will use a [convolutional neural network](http://cs231n.github.io/convolutional-networks/) backbone and a fully connected head with a single hidden layer as a classifier. Don't know what these things mean? Not to worry, we will dive deeper in the coming lessons. For the moment you need to know that we are building a model which will take images as input and will output the predicted probability for each of the categories (in this case, it will have 37 outputs).

We will train for 4 epochs (4 cycles through all our data).

```
# cnn_learner takes in a DataBunch and resnet model. there is a 34 and 50 version.
# to start, use the smaller resnet mode (34) because will run faster

# requires resnet34 download pre-trained weights. start with model that has been trained on various data
# transfer learning - take a model that can do something really well and can do your thing really well
# can take less time and less data to train data
learn = cnn_learner(data, models.resnet34, metrics=error_rate)

# validation set where model never saw and prevent overfitting
# fastai already built in for you
```

```
learn.model
```

```
# rather than use fit use fit_one_cycle
# more accurate and faster than previous approach!
# what is the difference?
# 4 is how many times do we go through the data. If model sees data multiple times, then overfit
# 4 epochs or cycles is a good start
# total of 1 min and 45s at error_rate of 6%  
learn.fit_one_cycle(4)
```

```
# save the weights of the model in the same directory but will separate between models, databunches
# save so you can still get this type of performance of 1:45 s
learn.save('stage-1')
```

```
# fastai more accurate than keras
# fastai has support for AWS and Google Cloud because code libraries on top of keras democratize data science
# resnet vs. inception?
# stanford dawn bench benchmark (dawn.cs.stanford.edu/benchmark) shows how well these architectures work
# edge computing (mobile phone) may need the other architectures but resnet works just fine
# inception is pretty memory intensive and not terribly resilient
```

## Results

Let's see what results we have got. 

We will first see which were the categories that the model most confused with one another. We will try to see if what the model predicted was reasonable or not. In this case the mistakes look reasonable (none of the mistakes seems obviously naive). This is an indicator that our classifier is working correctly. 

Furthermore, when we plot the confusion matrix, we can see that the distribution is heavily skewed: the model makes the same mistakes over and over again but it rarely confuses other categories. ## Unfreezing, fine-tuning, and learning rates

Since our model is working as we expect it to, we will *unfreeze* our model and train some more.

```
# make our model run better by fine-tuning
# our model ran really fast because deep learning NN have many layers and we added extra layers to end and only trained those.
# left the other layers as it
# NOW GO BACK AND TRAIN THE WHOLE MODEL
# unfreeze says please train whole model
learn.unfreeze()
```

```
# error got worse! WHY?
# visualize layers of CNN (RGB pixel values). resnet 34 has 34 layers
# visualize specific coefficients of layers. operates by group of pixels and finds diagonal lines in one direction or yellow to blue gradient
# layer 2 does another computation and look at something that looks for corners, and circles, etc.
# layer 1 does 1 line, layer 2 does 2 lines
# show what photos activated those layers -- some layers are good for finding window corners etc
# layer 3 can find repeating patterns or lines joined together or bits of text (horizontal pattern), geometric patterns
# layer 4 can find dog faces, bird legs
# layer 5 can find eyes, etc.
# when we first trained, we kept these layers. now finetune and change the above features/layers
# layer 1 is most basic and not really change much
# layer 5 looks at dog faces so maybe change this layer more. deeper layers represent more semantic complexity
# THIS CODE BELOW DID HORRIBLY BECAUSE CHANGE ALL LAYERS NOT JUST LAYER 5 (THE MOST COMPLEX)
learn.fit_one_cycle(1)
```

```
learn.load('stage-1');
```

```
# what is the fastest I can train this NN
learn.lr_find()
```

```
# plot the results of our LR finder (learning rate: how quickly am I updating numbers in model)
# loss gets worse after 10^(-1). default learning rate is 10^(-1), which is another reason why this is worse
learn.recorder.plot()
```

```
learn.unfreeze()
# pass range of learning rates using slice (takes stop and start values). first layers have 1e-6 and later layers have 1e-4
# not need all layers to have the new learning rate because the earlier layers were doing fine before
# good rule of thumb: pass last stage in learning rate slice to be 10x smaller than default(1e-3) and the start to be well before things go bad (1e-6)
learn.fit_one_cycle(2, max_lr=slice(1e-6,1e-4))
```

## Training: resnet50

```
# GPU run out of memory. more graphics card memory and more parameters to fit if run resnet50
# won't go over this part but run through and try it
# if 8GB of GPU memory, then error (need salamander or aws)
# fix using bs (batchsize). if run out of memory (have 8GB card), then reduce batchsize == 32. will just take longer
```

Now we will train in the same way as before but with one caveat: instead of using resnet34 as our backbone we will use resnet50 (resnet34 is a 34 layer residual network while resnet50 has 50 layers. It will be explained later in the course and you can learn the details in the [resnet paper](https://arxiv.org/pdf/1512.03385.pdf)).

Basically, resnet50 usually performs better because it is a deeper network with more parameters. Let's see if we can achieve a higher performance here. To help it along, let's us use larger images too, since that way the network can see more detail. We reduce the batch size a bit since otherwise this larger network will require more GPU memory.

```
data = ImageDataBunch.from_name_re(path_img, fnames, pat, ds_tfms=get_transforms(),
                                   size=299, bs=bs//2).normalize(imagenet_stats)
```

```
learn = cnn_learner(data, models.resnet50, metrics=error_rate)
```

```
learn.lr_find()
learn.recorder.plot()
```

```
learn.fit_one_cycle(8)
```

```
learn.save('stage-1-50')
```

It's astonishing that it's possible to recognize pet breeds so accurately! Let's see if full fine-tuning helps:

```
learn.unfreeze()
learn.fit_one_cycle(3, max_lr=slice(1e-6,1e-4))
```

If it doesn't, you can always go back to your previous model.

```
learn.load('stage-1-50');
```

```
interp = ClassificationInterpretation.from_learner(learn)
```

```
# helps to understand and become a domain expert because know that experts also struggle with some of these classifications
interp.most_confused(min_val=2)
```
